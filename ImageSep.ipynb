```python
import pandas as pd
import numpy as np
from PIL import Image, ImageSequence
```

```python
np.random.seed(10)
```

```python
testcsv = []
traincsv = []
centroidtrain = []
centroidtest = []

for i in range(2, 29):
    for j in range(1,9):
        try:
            data = pd.read_csv(f'C:/Users/hugom/Desktop/Analysis/cells/Image_{i:03}_0{j}-centroids.csv', header=None)
            data.iloc[:,1]=list(map(lambda x: 1 if x=='eGFP+' else 0, data.iloc[:,1]))
            trainbool = np.random.binomial(1, 0.8, len(data))
            for k in range(0, len(data)):
                if (trainbool[k]):
                    traincsv.append([f'Image_{i:03}_0{j}_{k+1}.png',data.iloc[k,1]])
                    centroidtrain.append([i, j, data.iloc[k,2], data.iloc[k,3]])
                    with Image.open(f'C:/Users/hugom/Desktop/Analysis/cells/Image_{i:03}_0{j}_{k+1}.tif') as im:
                        tmp = np.asarray(ImageSequence.all_frames(im)[1], dtype=np.uint16)[:64, :64]
                        im2 = Image.fromarray(tmp)
                        im2.save(f'./train/Image_{i:03}_0{j}_{k+1}.png')
                else:
                    testcsv.append([f'Image_{i:03}_0{j}_{k+1}.png', data.iloc[k,1]])
                    centroidtest.append([i, j, data.iloc[k,2], data.iloc[k,3]])
                    with Image.open(f'C:/Users/hugom/Desktop/Analysis/cells/Image_{i:03}_0{j}_{k+1}.tif') as im:
                        tmp = np.asarray(ImageSequence.all_frames(im)[1], dtype=np.uint16)[:64, :64]
                        im2 = Image.fromarray(tmp)
                        im2.save(f'./test/Image_{i:03}_0{j}_{k+1}.png')
            print(f'Image_{i:03}_0{j} done')
        except pd.errors.EmptyDataError:
            print(f'Image_{i:03}_0{j} not found')
        except FileNotFoundError:
            print(f'Image_{i:03}_0{j} not found')
```

```python
traincsv_pd = pd.DataFrame(traincsv)
traincsv_pd.to_csv('train.csv', index=False, header=False)
testcsv_pd = pd.DataFrame(testcsv)
testcsv_pd.to_csv('test.csv', index=False, header=False)
centroidtrain_pd = pd.DataFrame(centroidtrain)
centroidtrain_pd.to_csv('centroidtrain.csv', index=False, header=False)
centroidtest_pd = pd.DataFrame(centroidtest)
centroidtest_pd.to_csv('centroidtest.csv', index=False, header=False)
```
